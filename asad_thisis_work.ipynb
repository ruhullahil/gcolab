```
import pandas as pd

# Reading the dataset for english
df = pd.read_excel('data_dictionary.xlsx')

# Assign values for each of the word from the database
word_vals = dict()

for row, row_val in df.iterrows():
    # print(row_val['Good =1'])
    for col, (a, b) in enumerate(row_val.items()):
        if type(b) is float:
            continue
        a = float(a.split()[-1])
        b = b.strip().lower()
        word_vals[b] = a


# Process each sentence properly

def sentence_extract(line):
    temp = []
    for word in line.split():
        word = word.lower()
        if word == 'won\'t':
            temp.append('will')
            temp.append('not')
        elif word.endswith('n\'t'):
            temp.append(word[:-3])
            temp.append('not')
        elif word.endswith('\'ll'):
            temp.append(word[:-3])
            temp.append('will')
        elif word[-1] in ['.', ',']:
            temp.append(word[:-1])
        else:
            temp.append(word)
    return temp


# Calculates the positive and negative sum according to the algorithm
def calculate(words):
    negation, mixed, posSum, negSum = False, False, 0, 0

    for word in words:
        if word not in word_vals:
            continue
        if word_vals[word] == 1.5:
            mixed = True
            continue
        if word_vals[word] == -4:
            negation = True
            continue

        if mixed == True:
            val = word_vals[word] * 1.5
            mixed = False
        else:
            val = word_vals[word]

        if negation == True:
            val = word_vals[word] * -4
            negation = False
        else:
            val = word_vals[word]

        if val < 0:
            negSum += val * -1
        else:
            posSum += val

    return (posSum, negSum)


'''
n = int(input('Number of input sentences: '))
'''

testdata = pd.read_excel('input_data.xlsx')
n= len(testdata)
print(n)
'''
full =[]
for Row , Row_val in testdata.iterrows():
  Dic ={}
  for key,value in Row_val.items():
    #print(key ,value)
    Dic[key]=value
  full.append(Dic)

print(full)
    
'''
totPos, totNeg, tot = 0, 0, 0
lst =[]

for Row , Row_val in testdata.iterrows():
    posSum, negSum = calculate(sentence_extract(Row_val['review']))
    totSum = posSum + negSum
    totPos += posSum
    totNeg += negSum
    tot += totSum
   # print(f"{posSum},{negSum}")

    if posSum !=0:
      #print("test posetive")
      pos = round((posSum / totSum) * 100, 2)
    else :
      pos =0
    if negSum !=0:
      #print('test neg')
      neg = round((negSum / totSum) * 100, 2)
    else :
      neg = 0


      
    Dic ={}
    for key ,value in Row_val.items():
      Dic[key]=value
    Dic['postive']=pos
    Dic['negative']=neg
    print(f"band = {Row_val['feature']}")
    print(f"Positive ratio = {pos:0.2f}")
    print(f"Negative ratio = {neg:0.2f}")
    #print('\n\n')
    lst.append(Dic)
    Dic


print(lst)
wrt= pd.DataFrame(lst)
wrt.to_csv('output.csv')
posPercent = round((totPos/tot)*100, 2)
negPercent = round((totNeg/tot)*100, 2)

if posPercent >= 60:
    print("Excellent product")
elif posPercent >= 40:
    print("Good product")
elif posPercent >= 1:
    print("General product")
else:
    print("Bad product")
```
